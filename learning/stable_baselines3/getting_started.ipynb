# Getting Started: how to use Stable Baselines 3?
See: [https://stable-baselines3.readthedocs.io/en/master/guide/quickstart.html](https://stable-baselines3.readthedocs.io/en/master/guide/quickstart.html)

```python
import gymnasium as gym
from stable_baselines3 import A2C
```

```python
env = gym.make("CartPole-v1", render_mode="rgb_array")

model = A2C("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=10_000)
```

```python
vec_env = model.get_env()
obs = vec_env.reset()
for i in range(1000):
    action, _state = model.predict(obs, deterministic=True)
    vec_env.render() # This return an RGB array, need to display image using opencv or something else to see the result, but it's not really important
```
